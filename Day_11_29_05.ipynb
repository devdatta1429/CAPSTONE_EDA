## 🔶View of nulls, duplicate values of the DataSet🔶

```python
nulls = df.isnull().sum()
duplicated = df.duplicated().sum()
unique = df.nunique()
datatypes = df.dtypes

before_clean = pd.DataFrame({
    'Total Nulls' : nulls,
    'duplicated' : duplicated,
    'unique' : unique,
   'datatypes' : datatypes
})

before_clean
```

#
## 🔶Handling null value of return_reason,return_date🔶

```python
before = df[['return_reason' , 'return_date']].isnull().sum().to_frame(name='Missing_Count')
print(before)
print()

df['return_date'] = df['return_date'].fillna('Unknown')
df['return_reason'] = df['return_reason'].fillna('Unknown')

after = df[['return_reason' , 'return_date']].isnull().sum().to_frame(name='Missing_Count')
print(after)
after

```
